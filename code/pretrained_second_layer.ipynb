```python
def create_model_resnet():
  import tensorflow as tf
  from tensorflow import keras
  from tensorflow.keras import layers
  model = keras.Sequential([
      layers.Input(shape=(2048,)),  # Input layer with size 2048
      layers.Dense(512, activation='relu'),  # Hidden layer 1
      layers.Dense(256, activation='relu'),  # Hidden layer 2
      layers.Dense(128, activation='relu'),  # Hidden layer 3
      layers.Dense(10, activation='softmax')  # Output layer
  ])
  return model
```

```python
def create_model_cifar():
  import tensorflow as tf
  from tensorflow import keras
  from tensorflow.keras import layers
  model = keras.Sequential([
      layers.Input(shape=(512,)),  # Input layer with size 512
      layers.Dense(400, activation='relu'),  # Hidden layer 1
      layers.Dense(256, activation='relu'),  # Hidden layer 2
      layers.Dense(128, activation='relu'),  # Hidden layer 3
      layers.Dense(10, activation='softmax')  # Output layer
  ])
  return model
```

```python
def create_model_inception():
  import tensorflow as tf
  from tensorflow import keras
  from tensorflow.keras import layers
  model = keras.Sequential([
      layers.Input(shape=(2048,)),  # Input layer with size 2048
      layers.Dense(1024, activation='relu'),  # Hidden layer 1
      layers.Dense(512, activation='relu'),  # Hidden layer 2
      layers.Dense(256, activation='relu'),  # Hidden layer 3
      layers.Dense(128, activation='relu'),  # Hidden layer 4
      layers.Dense(10, activation='softmax')  # Output layer
  ])
  return model
```
